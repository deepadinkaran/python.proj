```python
#Explotary Data Analysis 
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Gender Distribution')
plt.show()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
#Age vs gender 

ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("age distribution")
plt.show()
```

```python
#total amount vs Age group 
sales_gn=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_gn)
```

```python
#gender+age group vs amount

sns.barplot(data=df,x='Age Group',y='Amount',hue='Gender')
for bars in ax.containers:
    ax.bar_labels(bars)
    sales_gn=df.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.show() 
plt.title("Age+Gender amt Distribution")
plt.xticks(rotation=45)
plt.show()
                    
```

```python
#occupation vs amount
sns.barplot(data=df,x='Occupation',y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.figsize(20,6)
    plt.xticks(rotation=45)
sales_gn=df.groupby(['Occupation'],as_index=False),['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.show()
plt.title("Buyers Profession")
plt.show()
```

```python
#State vs Amount 

for bars in ax.containers:
    ax.bar_label(bars)
    sales_st=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
ax= sns.barplot(data=df, x='State', y='Amount')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
#STATE
#top10 State + Orders

states_or=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(data=df, x='State', y='Orders')
sns.set(rc={'figure.figsize':(25,5)})
plt.show()
```

```python
#total amount /sales from top 10 State
state_totamt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by=['Amount'],ascending=False).head(10)
colors= ['blue','green','red','purple','orange','yellow','brown','pink','cyan','magenta']
sns.barplot(data=df, x='State', y='Amount',palette=colors)
plt.show()
```

```python
#maritial Status
ax=sns.countplot(data = df, x='Marital_Status')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Marital Distribution")
plt.tight_layout()
plt.show()
```

```python
#product
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pID=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_pID, x='Product_ID', y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
plt.tight_layout()
plt.show()
```
